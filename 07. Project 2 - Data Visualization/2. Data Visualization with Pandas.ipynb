```python
import pandas as pd 
```

```python
# read csv file 
df_population_raw = pd.read_csv('population_total.csv')
```

```python
# dataframe
df_population_raw
```

# 1 Making a Pivot Table

```python
# drop the null values 
df_population_raw.dropna(inplace=True)
```

```python
# Making a Pivot Table
df_pivot = df_population_raw.pivot(index='year', columns='country', values='population')
```

```python
# select some columns
df_pivot = df_pivot[['India', 'China', 'United States', 'Indonesia', 'United Kingdom', 'Brazil', 'Russia']]
```

```python
# show dataset
df_pivot
```

# 2 Lineplot

```python
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population',
             title='Population (1955-2020)', figsize=(6,4))
```

# 3 Bar Plot

### 3.1 Single Barplot

```python
# selecting only one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]

# df_pivot.index.isin([2020])
```

```python
# showing new dataframe
df_pivot_2020
```

```python
# transpose dataframe from rows to columns ir viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind='bar',color='red',
                  xlabel='Country', ylabel='Population',
                  title='Population (2020)', figsize=(6,4))
```

### 3.2 Barlpot grouped by "n" Variables

```python
# Selecting a few years 
df_pivot_years = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
# making grouped barplot
df_pivot_years.plot(kind='bar',
                  xlabel='Years', ylabel='Population',
                  title='Population (1980-2020)')
```

# 4. Piechart

```python
# changing column name
df_pivot_2020.rename(columns={2020: '2020'}, inplace=True)
# changing the Int 2020 to Str '2020'
```

```python
# making piechart
df_pivot_2020.plot(kind='pie', y='2020', title='Population in 2020(%)')

# y attribute : column that contain the data, here '2020' column contains pop data.
```

# 5 Boxplot

### 5.1 Single Boxplot

### 5.2 Multiple Boxplot

```python
df_pivot.plot(kind='box', ylabel='Population', color='purple')
```
