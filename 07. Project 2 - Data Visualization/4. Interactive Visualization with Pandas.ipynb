```python
# read the csv file 
df_population_raw = pd.read_csv('population_total.csv')

# dropping null values
df_population_raw.dropna(inplace=True)

# making a pivot table 
df_pivot = df_population_raw.pivot(index="year", columns="country", values="population")

# select some country columns 
df_pivot = df_pivot[['India', 'China', 'United States', 'Indonesia', 'United Kingdom', 'Brazil', 'Russia']]

# showing the pivot table
df_pivot
```

# 1 Lineplot  ( Interactive )

```python
# df_pivot.iplot(kind='line')
# cufflinks is now Deprecated/Unmaintained so NOT working
```

```python
# Using plotly ------ NEW AND WORKING
fig = px.line(df_pivot,
    x=df_pivot.index,
    y=df_pivot.columns,
    title="Population (1955-2020)")

# x is the index (years)
# y is columns (population)
# the lines are the countries. 
```

```python
# showing figure (Lineplot)
fig
```

# 2 Barplot (Interactive)

### 2.1 Barplot on Single value 

```python
# Selecting only one year (2020)
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
# Transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# df_pivot_2020.head()
df_pivot_2020.columns
# df_pivot_2020.index
```

```python
bar_fig = px.bar(df_pivot_2020, 
                 x=df_pivot_2020.index, 
                 y=df_pivot_2020.columns[0], 
                 title="Population (2020)")
```

```python
bar_fig.show()
```

### 2.2 Barplot Grouped by 'n' Variables

```python
# selecting a few years
df_pivot_group = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
df_pivot_group
```

```python
bar_group = px.bar(df_pivot_group, 
                 x=df_pivot_group.index, 
                 y=df_pivot_group.columns, 
                 title="Population (1980-2020)")
```

```python
bar_group.show()
```

# 3 Piechart

```python
# rename the column 
df_pivot_2020.rename(columns={2020:'2020'}, inplace=True)
```

```python
df_pivot_2020
```

```python
# reset index to make the index better and can be used to generate the piechart
df_2020 = df_pivot_2020.reset_index()
```

```python
pie_fig = px.pie(df_2020,
                 values='2020', names='country', title="Population 2020")
```
