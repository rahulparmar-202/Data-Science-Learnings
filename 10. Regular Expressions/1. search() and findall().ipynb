# RegEx

# The re Module 

```python
import re
```

## 1.1 re.search()

```python
text = """Hello World.
I love Python!!!!!
my first lucky number is 349-432-393
My second lucky number is 584-388-539
My third lucky number is 666-122-983-200-444
My fourth lucky number is 927-385-226-385
"""
```

```python
# return the first match 
re.search(r'\d', text)
```

```python
# flags - I -> Insensitive Flag 
re.search(r'^hello', text, flags=re.I)
```

## 2 re.findall()

```python
# returns all the matches 
re.findall(r'lucky', text)
```

## 3 Excercise 

```python
# Match the punctuation
text = """Hello World.
I love Python!!!!!
my first lucky number is 349-432-393
My second lucky number is 584-388-539
My third lucky number is 666-122-983-200-444
My fourth lucky number is 927-385-226-385
"""

re.findall(r'[^\w\s]', text)
```

```python
# match right date format (MM-DD-YYYY)
text = '''
13-04-2021
2021-13-04
2021-04-13
02-20-2005
'''
re.findall(r'\d{2}-\d{2}-\d{4}', text)
```

```python
# match right username format (4-14 characters and only letters or digits)
text = '''
username10
abc
10
'''

# \w -> Word character [a-zA-Z0-9_]
# {4, 14} -> only takes min 4 and max 14.

re.findall(r'[\w]{4,14}', text)
```
