```python
import pandas as pd 
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
df_laptops.head(3)
```

# 1. unique()

Returns the unique values of Series object. (Uniques are returned in the order of appearance)

```python
# get unique elements in the "Company" column
df_laptops['Company'].unique()
```

```python
# get unique elements in the "Inches" column.
df_laptops['Inches'].unique()
```

```python
# get the lengths of unique elements -- Inches
len(df_laptops['Inches'].unique())
```

```python
# get the lengths of unique elements -- Company
len(df_laptops['Company'].unique())
```

# 2. nunique()

return number of unique elements in the object (excludes NA values by default)

```python
# get number of unique elements in the "Company" column
df_laptops['Company'].nunique()
```

```python
# get number of unique elements in the "Inches" column
df_laptops['Inches'].nunique()
```
