```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv',encoding='latin1')
```

```python
df_laptops.head(3)
```

# 1 isin() : Single Filtering

```python
# Select Apple or HP laptops
df_laptops['Company'].isin(["Apple", "HP"])
```

```python
# Filter the dataframe 
df_laptops[df_laptops['Company'].isin(["Apple", "HP"])].value_counts('Company')
```

# 2 isin() : Multiple Filtering

```python
# Find notebooks or ultrabooks from Apple or HP
filter1 = df_laptops['TypeName'].isin(["Ultrabook", "Notebook"])
filter2 = df_laptops['Company'].isin(["Apple", "HP"])
```

```python
# filter dataframe
df_laptops[filter1 & filter2].value_counts(['Company', 'TypeName'])
```
