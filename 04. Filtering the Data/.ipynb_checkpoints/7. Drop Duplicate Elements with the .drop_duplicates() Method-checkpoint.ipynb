```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
df_laptops.head(3)
```

# 1 .drop_duplicates()

```python
# drop duplicates in 2 or more column
df_laptops.drop_duplicates('Company')[['Company','Price_euros']]
```

```python
# sort dataframe in ascending by company and price
# cheapest first and most expensive last
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
# cheapest keep='first' -------- this means it only keeps the first value of each duplicate company *
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# most expensive keep='last'
df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
```

 -- **inplace** : drop duplicates in place or to return a copy,
 
 -- **ignore_index=True** : resulting axes will be labeled as 0,1,...

```python
# Arguments 

# keep last will only keep the last of each duplicates---since the df is sorted so the last of each duplicate will be exepensive **
df_laptops[['Company', 'Price_euros']]
```

# 2 Excercise 1

Task: Find the biggest and smallest screen size in laptops of each company
using the **sort_values()** and **duplicated** methods (keep "first","last" and False)

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
# sort dataframe ascending by Company and Inches
# smallest first and biggest last
df_laptops = df_laptops.sort_values(['Company', 'Inches'])
```

```python
# smallest : keep='first'
df_laptops_first = df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Inches']]
df_laptops_first
```

```python
# biggest : keep='last'
df_laptops_last = df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Inches']]
df_laptops_last
```

### Counting Values ------------------

```python
# keep = first
df_laptops_first.value_counts(['Company', 'Inches'])
```

```python
# keep = last

df_laptops_last.value_counts(['Company', 'Inches'])
```
