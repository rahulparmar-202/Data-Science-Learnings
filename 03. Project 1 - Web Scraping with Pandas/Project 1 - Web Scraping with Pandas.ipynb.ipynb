# 2 read .csv from Multiple URLs with Pandas

- https://www.football-data.co.uk/mmz4281/2425/E0.csv
- https://www.football-data.co.uk/mmz4281/2425/E1.csv

**LINK = root + season + league**

```python
# Link Structure
"https://www.football-data.co.uk/mmz4281/" + "2425" + "/" + "E0" + ".csv"
```

```python
# Create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiple Leagues

- https://www.football-data.co.uk/mmz4281/2425/E0.csv
- https://www.football-data.co.uk/mmz4281/2425/E1.csv
- https://www.football-data.co.uk/mmz4281/2425/E2.csv
- https://www.football-data.co.uk/mmz4281/2425/E3.csv
- https://www.football-data.co.uk/mmz4281/2425/EC.csv

```python
# Creating a list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

# Looping through the Leagues, read multiple CSVs and append it into a list
for league in leagues:
    df = pd.read_csv(root + '2425' + '/' + league + '.csv')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

## 2.2 Multiple Seasons

```python
# Creating a list of leagues
leagues = ['E0', 'E1', 'E2']
frames = []

# Loop through the multiple seasons and insert a new column

for league in leagues:
    for season in range(19, 25):
        df = pd.read_csv(root + str(season)+str(season+1) + '/' + league + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# Total Frames (6 Seasons x 3 Leagues)
len(frames)
```

```python
# show 1st and last frames, (new column was created)
frames[0]
frames[17]
```

# 3 Organize All the Data in a Dictionary

```python
# Create a dictionary with original names of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}
# Create List of leagues
# leagues = ['E0', 'E1', 'E2']


# Loop through the multiple seasons and insert a new column
for league in dict_countries:
    frames = []
    for season in range(19, 25):
        df = pd.read_csv(root + str(season)+str(season+1) + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
# show dataFrame inside dict_historical_data
dict_historical_data['Spanish La Liga']
```
