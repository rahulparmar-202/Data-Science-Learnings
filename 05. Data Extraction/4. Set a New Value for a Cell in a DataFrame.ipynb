```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select columns 
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df
```

# 1. Set value to one cell 

```python
# update L. Messi's height

# height before was = 170 cm 
df.loc['L. Messi', 'height_cm'] = 172
```

```python
# show dataframe
df.head()
```

# 2. Set Value to entire Column

```python
# set the height of all players to 190

# to see the height of all players 
# df.loc[:,'height_cm']  

# update :
df.loc[:, 'height_cm'] = 190
```

```python
# show dataframe
df
```

# 3. Set Value to entire Row

```python
# get all columns that corresponds to the players ranked last in fifa 
df.iloc[-1, :]
```

```python
# set NULL values to all the players that are ranked last in fifa
import numpy as np

df.iloc[-1, :] = np.nan
```

```python
# show dataframe
df
```

# 4. Set values to Multiple cells

```python
# Set value for all items matching the lists of labels.
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 175
```

```python
# show dataframe 
df
```

# 5. Set value for rows matching a condition

```python
# set value for row matching a condition
columns = ['age', 'dob', 'height_cm', 'weight_kg']

df.loc[df['height_cm']>180, columns] = 0
```

```python
# show dataframe 
df
```
