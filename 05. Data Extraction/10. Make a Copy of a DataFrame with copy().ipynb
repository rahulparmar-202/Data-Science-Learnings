```python
import pandas as pd
```

```python
# read csv
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select columns
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

# 1. copy()

## 1.1 deep=True

```python
# deep=True by dedault (Modifications to the data or indices of the copy)
df_copy = df.copy()

# here we created a copy of df that is independant Now --------
```

```python
# update value in original dataframe
df.loc['L. Messi', 'height_cm'] = 180
# only getting updated in the original dataframe
```

```python
# copy vs original dataframe
df
df_copy
```

## 1.2 deep=False 

```python
# deep=False  (Any changes to the DataFrame will affect the Original and the shalow copy also.)
df_cp = df.copy(deep=False)
```

```python
# update value in the Original DataFrame
df.loc['L. Messi', 'height_cm'] = 175
# height was 180 (last updated)
```

```python
# shalow copy VS original df
# df_cp.loc['L. Messi', 'height_cm'] # 175
df.loc['L. Messi', 'height_cm']  # 175
```

```python
# both are updated the same 
```
