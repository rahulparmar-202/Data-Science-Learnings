```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select columns
df = df[['long_name', 'age',	'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
df.head()
```

## 1. Selecting with Single Column 

loc[row_label, column_label]

```python
# get all data about L.Messi
df.loc['L. Messi']
```

```python
# get the height of L. Messi
df.loc['L. Messi','height_cm']
```

```python
# get the weight of Cristiano Ronaldo
df.loc['Cristiano Ronaldo','weight_kg']
```

```python
# get all the rows inside height_cm column
df.loc[:,'height_cm']
```

```python
# get all the colums that corresponds to the index L. Messi index
df.loc['L. Messi',:]
```

## 2. Selecting with a list of Values 

```python
# get all the data about L. Messi and Crisitiano Ronaldo 
df.loc[['L. Messi', 'Cristiano Ronaldo']]
```

```python
# get the height_cm of L.Messi and Ronaldo
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']]
```

```python
# get the height, weight of L. Messi
df.loc[['L. Messi'], ['height_cm', 'weight_kg']]
```

```python
# get the height and weight of L.Messi and Ronaldo
df.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm', 'weight_kg']]
```

## 3. Selecting a Range of Data with Slice

start:stop:step , NOTE:- Contrary to usual python slice , this slice include both start and stop (inclusive).

```python
# slice column labels 
players = ['L. Messi', 'Cristiano Ronaldo']

df.loc[players, 'age':'club']
```

```python
# slice index labels
columns = ['age','dob', 'height_cm', 'weight_kg', 'nationality', 'club']

# get top1 and top10 players
df.index[:10] # to get the top10 players names 

df.loc['L. Messi':'M. Salah', columns]
```

## 4. Selecting with Conditions

```python
# one condition: select players with height more than 180 cm 
columns = ['age','dob', 'height_cm', 'weight_kg']

df.loc[df['height_cm'] > 180, columns]
```

```python
# Multiple Condition: select players with height more than 180cm and from Argentina
columns = ['age','dob', 'height_cm', 'weight_kg', 'nationality', 'club']

df.loc[(df['height_cm'] > 180) & (df['nationality']=='Argentina'), columns]
```
