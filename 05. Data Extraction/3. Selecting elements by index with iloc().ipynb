```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select columns
df = df[['long_name', 'age',	'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe 
df
```

# 1. Selecting with a Single Value

iloc[ row_posi, column_posi ]

```python
# get the height of L. Messi

# both index starts from 0
# first row = index 0
# height_cm column index = 3

df.iloc[0, 3]
```

```python
# get the weight of Cristiano Ronaldo

df.iloc[1, 4]
```

```python
# get all rows inside "height_cm" column

df.iloc[:, 3]
```

```python
# get all the columns that correspond to 'L. Messi'

df.iloc[0, :]
```

# 2. Selecting with a list of values

```python
# get all data about Leonal messi and Cristiano Ronaldo 
df.iloc[[0, 1], :]
```

```python
# get the height of L messi and Cristiano Ronaldo
df.iloc[[0,1], [3]]
```

```python
# get the height and weight of L. Messi
df.iloc[[0], [3,4]]
```

```python
# get the height and weight of L. Messi & Cristiano Ronaldo 
df.iloc[[0,1], [3,4]]
```

# 3. Selecting a Range of data with Slice

start:stop (NOTE: start is included but the stop is excluded)

```python
# Slice column lables : age/club
# players = ['L. Messi', 'Cristiano Ronaldo']
players = [0, 1]
df.iloc[players, 1:7]
```

```python
# slice index labels (top1 to top10 players)
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1,2,3,4]

# get top1 and top10 player name
df.iloc[0:10, columns]
```

# 4. Selecring with Conditions

iloc() cannot accept a boolean , but boolean list. We have to use the list() function to convert a Series into a list.

```python
# one condition : selecting one player with height above 180cm 
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1,2,3,4]

df.iloc[list(df['height_cm']>180), columns]
```

```python
# multiple condition: select a player with height more than 180cm and from Argentina

# columns = ['age', 'dob', 'height_cm', 'weight_kg', 'nationality']
columns = [1,2,3,4,5]

df.iloc[list( (df['height_cm']>180) & (df['nationality']=='Argentina') ), columns]
```
