```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select columns 
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe 
df
```

# 1. sample()

```python
# extract 10 random elements from 'nationality' column
df['nationality'].sample(10, random_state=99)

# 99 here is the key to this same random elements 
```

```python
# extract a random 20% sample of the dataframe
# df.sample(18278/5)   # --- that is 20% of 18278 rows

# 20%
df.sample(frac=0.2)
```

```python
# upsample: increase the sampling rate -> frac>1 (NOTE:- replace parameter has to be True for frac parameter > 1)
df.sample(frac=2, replace=True, random_state=56)

# 56 here is the key to the random elements 
```
