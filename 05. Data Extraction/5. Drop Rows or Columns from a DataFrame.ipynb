```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')

# set index 
df.set_index('short_name', inplace=True)

# select columns 
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

# 1. Drop Row

```python
# drop one row -----------

# axis parameter 
# df.drop('L. Messi', axis=0)

# index parameter
df.drop(index=['L. Messi'])
```

```python
# drop two or more rows and update the data (inplace=True)

# inplace=True :- means the original dataframe would be updated.

df.drop(index=['L. Messi', 'Cristiano Ronaldo'], inplace=True)

# df.drop(['L. Messi', 'Cristiano Ronaldo'], axis=0, inplace=True)
```

```python
df
```

# 2. Drop Columns

```python
# drop one column 
# axis parameter - 'age' column
df.drop('age', axis=1)

# columns parameter
df.drop(columns=['long_name'])
```

```python
# drop column by position\index (drop last column)
# df.columns[[-1]] :- to get to know the last column

df.drop(df.columns[[-1]], axis=1)
```

```python
# drop two or more columns and update data (inplace=True)
# columns parameter ---
# df.drop(columns=['dob', 'long_name'])

# axis parameter ---
df.drop(['dob', 'long_name'], axis=1, inplace=True)
```
