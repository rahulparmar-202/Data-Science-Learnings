```python
import pandas as pd
import numpy as np
```

## 1. Creating a DataFrame from an Array

### 1.1 Option-1

```python
# creating an array
data = np.array([[1,3], [2,4], [6,8], [5,10]])
```

```python
# creating a data frame 
indexes = ['row1', 'row2', 'row3','row4']
columns = ['col1', 'col2']
data_df = pd.DataFrame(data, index=indexes,columns=columns)
```

```python
# showing the data frame
data_df
```

### 1.2 Option-2

```python
# Creating an array with list shape
data = [[11,13], [12,14], [16,18]]
```

```python
# creating a data frame 
indexes = ['row1', 'row2', 'row3']
columns = ['col1', 'col2']
data_df = pd.DataFrame(data, index=indexes, columns=columns)
```

```python
# showing the data frame
data_df
```

## 2. Creating a DataFrame from a dictionary

```python
# lists used for this Example
cities = ["Tokyo","Yokohama","Osaka","Nagoya"]
population = [14195730, 3772726, 2788758, 2331078]
```

```python
# storing lists within the dictionary
data_dict = {'Cities': cities, 'Population': population}
```

```python
# creating DataFrame
df = pd.DataFrame(data_dict)
```

```python
df
```
