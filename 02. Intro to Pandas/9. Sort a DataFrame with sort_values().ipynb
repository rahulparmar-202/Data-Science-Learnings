```python
import pandas as pd
```

```python
# read csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams
```

# 1. Sort a DataFrame

```python
# Sort a column 
df_exams.sort_values(by='math score')
```

```python
# sort descending by one column
df_exams.sort_values('math score', ascending=False)
```

```python
# sort descending by multiple column
df_exams.sort_values(['math score', 'reading score', 'writing score'], ascending=False)
```

```python
# sort descending by multiple column and update the DataFrame
df_exams.sort_values(['math score', 'reading score', 'writing score'], ascending=False, inplace=True)
df_exams
```

```python
# also we can just over-ride the previouse DataFrame -> df_exams = "---"
```

```python
# Sort descending with a key function
df_exams.sort_values('race/ethnicity', ascending=True, key=lambda col:col.str.lower())
```
