### 1.2 Adding new column with Array

The DataFrame has 1000 rows so we need to create an array with 1000 elements.

```python
# import numpy
import numpy as np
```

```python
# create an array of 1000 elements
language_score = np.arange(0,1000)
```

```python
# len of language_score array
len(language_score)
```

```python
# update the language_score in the DataFrame
df_exams['language score'] = language_score
df_exams
```

```python
# create random integers between 1 to 100 
int_language_score = np.random.randint(1,100, size=1000)
# size is how many random integers to create...
```

In `random.randint(a, b)` : *a* is **inclusive** and *b* is **exclusive** (not included), So here 100 is not included.

```python
# check the len of the new random score array
len(int_language_score)
```

```python
# update the DataFrane column 'language_score' with the new score (random values 1-99)
df_exams['language score'] = int_language_score
```

```python
df_exams
```

```python
# create random float numbers between 1 to 100
np.random.uniform(1, 100, size=1000)
```

The `uniform(a, b, size)` : returns random **float** number between *a and b* (b not *included*), size is how many value we want.
