```python
import pandas as pd
```

```python
# read and show the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams
```

# 1 Create Index

```python
import numpy as np
import random
```

```python
# Creating non-repititive values for index
new_index = np.arange(0,1000)
```

```python
# shuffling the index
random.shuffle(new_index)
```

```python
# create new column with a new index
df_exams['new_index'] = new_index
df_exams
```

# 2 Set Index

```python
# set new_index column as index
df_exams.set_index('new_index', inplace=True)
```

```python
df_exams
```

```python
# Sort ascending by Index
df_exams.sort_index()
```

```python
# sort descending by index and save 
df_exams.sort_index(ascending=False, inplace=True)
```
