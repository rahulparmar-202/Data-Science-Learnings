```python
import pandas as pd
```

```python
df_movies = pd.read_csv('IMDB movies.csv', low_memory=False)
df_ratings = pd.read_csv('IMDB ratings.csv')
```

```python
df_movies = df_movies[['imdb_title_id', 'title', 'year', 'genre', 'country']]
df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

# 1 merge()

```python
# merge on 'id' :- it only counts the common columns and data. 
# that is available to both the dataframes.
```

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                   'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['C', 'D', 'E', 'F'],
                   'job': ['Doctor', 'Statistician', 'Accountant', 'Developer']})
```

## 1.1 Inner Join 

```python
df_in_join = df1.merge(df2, on='id', how='inner')
```

```python
print(df1.shape)
print(df2.shape)
print(df_in_join.shape)
```

```python
df_in_join
```

# 1.3 Exclusive Outer join (Exclusive Full Join)

combines rows and columns that are not commons.

```python
# show dataframes 
print(df1)
print("")
print(df2)
```

```python
# normal outer join 
df1.merge(df2, on='id', how='outer')
```

```python
# indicator = 'True'
df1.merge(df2, on='id', how='outer', indicator=True)
```

```python
# query --- (left only and Right only) NOT COMMON
df1.merge(df2, on='id', how='outer', 
          indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

### 1.3.1 Excercise

```python
# merge df_movies and df_ratings 
df_movies.merge(df_ratings, on='imdb_title_id', 
                how='outer', indicator=True)
```

```python
# Exclusive Outer Join 
df_movies.merge(df_ratings, on='imdb_title_id', how='outer', 
                indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

```python
# there are no data that is not common.
```

# 1.5 Exclusive Left Join

contains only the data that belongs to left side DataFrame.

```python
# shape 
print(df1.shape)
print(df2.shape)
```

```python
# merge Exclusive Left ----- indicator=True, and .query()
df_left = df1.merge(df2, on='id', how='outer', indicator=True).query("_merge=='left_only'")
```

```python
df_left.shape
```

```python
df_left
```

### 1.5.1 Excercise 

```python
# make a copy of the df_movies dataFrame
df_movies2 = df_movies.copy()
```

```python
df_movies2.index
```

```python
# set the first 1000 values of 'imdb_title_id' column as 'tt1234567890'
for index in df_movies2.index:
    if index < 1000:
        df_movies2.loc[index, 'imdb_title_id'] = "tt1234567890"
```

```python
# merge df_movies_2 and df_ratings (exclusive left join)
df_left_exclusive = df_movies2.merge(df_ratings, on='imdb_title_id', how='outer', 
                  indicator=True).query("_merge=='left_only'")
```

```python
# shape
print(df_left_exclusive.shape)
print(df_ratings.shape)
print(df_movies2.shape)
```

# 1.6 Right Join

# 1.7 Exclusive Right Join
